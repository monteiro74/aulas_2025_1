# Análise de Rede com Tipos de Interação e Pesos
Este exemplo mostra como modelar uma rede social simulada usando tipos de interação (menção, curtida, retweet) com pesos distintos para cada uma.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Criar grafo direcionado com pesos
G = nx.DiGraph()
interacoes_pesadas = [
    ("Ana", "Bruno", "menção", 1),
    ("Bruno", "Carlos", "retweet", 3),
    ("Carlos", "Diana", "curtida", 2),
    ("Diana", "Ana", "menção", 1),
    ("Ana", "Carlos", "curtida", 2),
    ("NewsBot", "Ana", "retweet", 3),
    ("NewsBot", "Bruno", "menção", 1),
]
for origem, destino, tipo, peso in interacoes_pesadas:
    G.add_edge(origem, destino, tipo=tipo, peso=peso)
```

```python
# Layout e visualização
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=1000)
nx.draw_networkx_edges(
    G, pos, arrowstyle='->', arrowsize=15,
    width=[G[u][v]['peso'] for u, v in G.edges()]
)
nx.draw_networkx_labels(G, pos, font_size=10)
edge_labels = {
    (u, v): f"{G[u][v]['tipo']} ({G[u][v]['peso']})" for u, v in G.edges()
}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='darkgreen')
plt.title("Rede com Tipos de Interação e Pesos")
plt.axis('off')
plt.tight_layout()
plt.show()
```
